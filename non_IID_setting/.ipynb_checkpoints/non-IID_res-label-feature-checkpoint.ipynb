```python
def get_data():

    columns=['Dataset','nr_clients','acc', 'lr', 'mom', 'bs','gt_lr','gt_mom','gt_bs','gt_acc']
    df = pd.DataFrame(columns=columns)
#
#     We dont have ground-truth batchsize results so those are just random!!!
#
    df = pd.concat([df, pd.DataFrame([['MNIST', '10 Clients', 0.9845, 0.5,0.6,14,0.5,0.9,10,0.9880]], columns=columns)])
    df = pd.concat([df, pd.DataFrame([['MNIST', '20 Clients', 0.9803, 0.5,0.5999999999999999,13,0.5,0.9,10,0.9875]], columns=columns)])

    df = pd.concat([df, pd.DataFrame([['EMNIST', '10 Clients', 0.8526, 0.5,0.3,74,0.5,0.6,10,0.8639]], columns=columns)])
    df = pd.concat([df, pd.DataFrame([['EMNIST', '20 Clients', 0.8458, 0.5,0.5999999999999999,96,0.5,0.6,10,0.8645]], columns=columns)])

    df = pd.concat([df, pd.DataFrame([['SVHN', '10 Clients', 0.8353, 0.5,0.3,11,0.5,0.6,10,0.8804]], columns=columns)])
    df = pd.concat([df, pd.DataFrame([['SVHN', '20 Clients', 0.8312, 0.5,0.6,11,0.5,0.6,10,0.8729]], columns=columns)])

    df = pd.concat([df, pd.DataFrame([['CIFAR-10', '10 Clients', 0.5811, 0.5,0.5999999999999999,34,0.5,0.9,10,0.6596]], columns=columns)])
    df = pd.concat([df, pd.DataFrame([['CIFAR-10', '20 Clients', 0.4737, 0.5,0.6,18,0.5,0.9,10,0.5731]], columns=columns)])


    return df

def plot_barplot(df, hue=[], title='', saveTitle='', saveFig=True, savePath='results/',param='', param_ann='',gt=''):
    sns.set_theme()
    plt.figure(figsize=(10, 5))

    ax = sns.barplot(y=param_ann, x="Dataset", hue="nr_clients",
                hue_order=hue, 
                data=df)
    # df['lr'].plot(y=param_ann, marker='*', color='black', ms=10)
    # ax.scatter(df['nr_clients'], df['lr'])
    handles, labels = ax.get_legend_handles_labels()

    sns.stripplot(
        data=df,
        x="Dataset",
        y=gt,
        hue="nr_clients",
        dodge=True,
        edgecolor="black",
        color = "black",
#         linewidth=.9,
        size=5.5,
        ax=ax
    )
    plt.legend(loc='lower left')
    plt.setp(ax.get_legend().get_texts(), fontsize='20')
    plt.title(title, fontsize='24')

    ax.xaxis.label.set_fontsize(24)
    ax.yaxis.label.set_fontsize(24)
    ax.tick_params(labelsize=24)
    plt.ylabel(param)
    plt.tight_layout()
    plt.ylim([0, 1])

    handles, labels = ax.get_legend_handles_labels()
    labels=['Ground Truth', 'Ground Truth', '10 Clients', '20 Clients']
    l = plt.legend(handles[1:], labels[1:], loc='lower left')
    plt.setp(ax.get_legend().get_texts(), fontsize='20')

    if saveFig:
        plt.savefig(param+".pdf")

    plt.show()
```

```python
df = get_data()
hue = ['10 Clients', '20 Clients']
plot_barplot(df, title="Learning Rate - Non-IID Setting", saveFig=True, hue=hue,param="Learning Rate",param_ann="lr", gt="gt_lr")
```

```python
df = get_data()
hue = ['10 Clients', '20 Clients']
plot_barplot(df, title="Momentum - Non-IID Setting", saveFig=True, hue=hue,param="Momentum",param_ann="mom", gt="gt_mom")
```

```python
df = get_data()
hue = ['10 Clients', '20 Clients']
plot_barplot(df, title="Accuracy - Non-IID Setting", saveFig=True, hue=hue,param="Accuracy",param_ann="acc", gt="gt_acc")
```
