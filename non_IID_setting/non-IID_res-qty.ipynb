```python
import sys
sys.path.append('../../')

import json
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# Set the font size for x tick labels
plt.rc('xtick', labelsize=30)
# Set the font size for y tick labels
plt.rc('ytick', labelsize=30)
from matplotlib.lines import Line2D

```

```python
def get_data():

    columns=['Dataset','nr_clients','acc', 'lr', 'mom', 'bs','gt_lr','gt_mom','gt_bs','gt_acc']
    df = pd.DataFrame(columns=columns)
#
#     We dont have ground-truth batchsize results so those are just random!!!
#
    df = pd.concat([df, pd.DataFrame([['MNIST', '10 Clients', 0.9870, 0.5,0.6,14,0.3,0.6,10,0.9884]], columns=columns)])
    df = pd.concat([df, pd.DataFrame([['MNIST', '20 Clients', 0.9823, 0.5,0.6,13,0.5,0.6,10,0.9862]], columns=columns)])

    df = pd.concat([df, pd.DataFrame([['EMNIST', '10 Clients', 0.8596, 0.5,0.6,74,0.5,0.3,10,0.8626]], columns=columns)])
    df = pd.concat([df, pd.DataFrame([['EMNIST', '20 Clients', 0.8557, 0.5,0.6,96,0.5,0.6,10,0.8589]], columns=columns)])

    df = pd.concat([df, pd.DataFrame([['SVHN', '10 Clients', 0.8716, 0.5,0.6,11,0.5,0.6,10,0.8804]], columns=columns)])
    df = pd.concat([df, pd.DataFrame([['SVHN', '20 Clients', 0.8576, 0.5,0.6,11,0.5,0.6,10,0.8729]], columns=columns)])

    df = pd.concat([df, pd.DataFrame([['CIFAR-10', '10 Clients', 0.6430, 0.5,0.6,34,0.5,0.6,10,0.6732]], columns=columns)])
    df = pd.concat([df, pd.DataFrame([['CIFAR-10', '20 Clients', 0.6425, 0.5,0.3,18,0.5,0.6,10,0.6425]], columns=columns)])


    return df

def plot_barplot(df, hue=[], title='', saveTitle='', saveFig=True, savePath='results/',param='', param_ann='',gt=''):
    sns.set_theme()
    plt.figure(figsize=(10, 5))

    ax = sns.barplot(y=param_ann, x="Dataset", hue="nr_clients",
                hue_order=hue, 
                data=df)
    # df['lr'].plot(y=param_ann, marker='*', color='black', ms=10)
    # ax.scatter(df['nr_clients'], df['lr'])
    handles, labels = ax.get_legend_handles_labels()

    sns.stripplot(
        data=df,
        x="Dataset",
        y=gt,
        hue="nr_clients",
        dodge=True,
        edgecolor="black",
        color = "black",
#         linewidth=.9,
        size=5.5,
        ax=ax
    )
    plt.legend(loc='lower left')
    plt.setp(ax.get_legend().get_texts(), fontsize='20')
    plt.title(title, fontsize='24')

    ax.xaxis.label.set_fontsize(24)
    ax.yaxis.label.set_fontsize(24)
    ax.tick_params(labelsize=24)
    plt.ylabel(param)
    plt.tight_layout()
    plt.ylim([0, 1])

    handles, labels = ax.get_legend_handles_labels()
    labels=['Ground Truth', 'Ground Truth', '10 Clients', '20 Clients']
    l = plt.legend(handles[1:], labels[1:], loc='lower left')
    plt.setp(ax.get_legend().get_texts(), fontsize='20')

    if saveFig:
        plt.savefig(param+".pdf")

    plt.show()
```

```python
df = get_data()
hue = ['10 Clients', '20 Clients']
plot_barplot(df, title="Learning Rate - Non-IID Setting", saveFig=True, hue=hue,param="Learning Rate",param_ann="lr", gt="gt_lr")
```

```python
df = get_data()
hue = ['10 Clients', '20 Clients']
plot_barplot(df, title="Momentum - Non-IID Setting", saveFig=True, hue=hue,param="Momentum",param_ann="mom", gt="gt_mom")
```

```python
df = get_data()
hue = ['10 Clients', '20 Clients']
plot_barplot(df, title="Accuracy - Non-IID Setting", saveFig=True, hue=hue,param="Accuracy",param_ann="acc", gt="gt_acc")
```
